**BRITISH MUSEUM SEARCH**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with 8 nodes
G = nx.Graph()

nodes=['S','A','B','C','E','D','G']
G.add_nodes_from(nodes)

# Add edges between the nodes
edges = [('S', 'A'), ('S', 'B'), ('A', 'D'), ('A', 'B'), ('B', 'C'), ('C', 'E'), ('D', 'G')]
G.add_edges_from(edges)

# Set the starting node (by label)
starting_node = 'S'

# Plot the graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_nodes(G, pos, nodelist=[starting_node], node_color='red')
plt.show()
```

**DEPTH FIRST SEARCH**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with 8 nodes
G = nx.Graph()

nodes=['S','A','B','C','E','D','G']
G.add_nodes_from(nodes)

# Add edges between the nodes
edges = [('S', 'A'), ('S', 'B'), ('A', 'D'), ('B', 'C'), ('C', 'E'), ('D', 'G')]
G.add_edges_from(edges)

# Perform Depth First Search
dfs_edges = list(nx.dfs_edges(G, source='S'))

# Set the starting node (by label)
starting_node = 'S'

# Plot the graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_nodes(G, pos, nodelist=[starting_node], node_color='red')

# Draw DFS edges in red
nx.draw_networkx_edges(G, pos, edgelist=dfs_edges, edge_color='red')

plt.show()
```

**BREADTH FIRST SEARCH**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with 8 nodes
G = nx.Graph()
nodes = ['S', 'A', 'B', 'C', 'E', 'D', 'G']
G.add_nodes_from(nodes)

# Add edges between the nodes
edges = [('S', 'A'), ('S', 'B'), ('A', 'D'), ('A', 'B'), ('B', 'C'), ('C', 'E'), ('D', 'G')]
G.add_edges_from(edges)

# Set the starting node (by label)
starting_node = 'S'

# Perform BFS
visited = set()
queue = [(starting_node, None)]  # Using a tuple to keep track of parent node
bfs_edges = []

while queue:
    node, parent = queue.pop(0)
    if node not in visited:
        visited.add(node)
        if parent is not None:
            bfs_edges.append((parent, node))
        for neighbor in G.neighbors(node):
            if neighbor not in visited:
                queue.append((neighbor, node))

# Create a new graph for the BFS tree
BFS_tree = nx.Graph()
BFS_tree.add_edges_from(bfs_edges)

# Plot the BFS tree
pos = nx.spring_layout(BFS_tree)  # Layout for the nodes
nx.draw(BFS_tree, pos, with_labels=True, node_color='lightblue', node_size=1500)
plt.show()
```
